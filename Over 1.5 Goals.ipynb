```python
import os
import pandas as pd
from datetime import datetime
# for inline plots in jupyter
%matplotlib inline
# import matplotlib
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(16,8)})
```

```python
data_dir = 'english-premier-league_zip/data'
files = os.listdir(data_dir)
files = [os.path.join(data_dir, i) for i in files if i[-4:]=='.csv']
```

```python
seasons = []
for file in files:
    season_year = file[-12:-8]
    season = pd.read_csv(file)
    season['Date'] = pd.to_datetime(season['Date'])
    season['Season'] = season_year
    seasons.append(season)
    
df = pd.concat(seasons).reset_index(drop=True)
```

Question: 

How many times do all matches have at least two goals on any given Saturday?

```python
satDateGoals = sat.groupby(['Season', 'Date'])['Goals'].agg({
    'Goals': [lambda x: (x > 2.5).sum(), 'count']
})
satDateGoals.columns = [' '.join(i) for i in satDateGoals.columns]
satDateGoals = satDateGoals.rename(
    columns={
        'Goals <lambda>': 'over 2.5',
        'Goals count': 'Matches'
    }
)
```

```python
satDateGoals['over Success'] = satDateGoals['over 2.5'] / satDateGoals['Matches']
```

```python
match_days = satDateGoals[satDateGoals['Matches'] > 4]
```

```python
match_days['over Success'].mean()
```

```python
wager = 10
number_of_bets = 3
winning_probability = match_days['over Success'].mean()
return_odds = 6
winning_bets = int(number_of_bets * winning_probability)
losing_bets = int(number_of_bets - winning_bets)
print(winning_bets*wager*return_odds - losing_bets*wager)
```
